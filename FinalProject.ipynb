```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('mobile data set.csv')
data.describe()
```

```python
x= data.drop('price_range', axis=1)
y= data['price_range']
```

Data Visualisation

```python
plt.scatter(x['pc'], x['ram'], c=y, cmap='viridis')
plt.xlabel('Pc')
plt.ylabel('ram')
plt.colorbar(label='Price Range')
plt.title('Scatter Plot: Price Range vs. ram')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
ss=StandardScaler()
#feature engneering
x_train=ss.fit_transform(x_train)
x_test=ss.fit_transform(x_test)
#renaming the target variables
l=LabelEncoder()
y=l.fit_transform(y)
```

```python
from sklearn.svm import SVC
svm_model = SVC(kernel='linear')
svm_model.fit(x_train, y_train)
y_pred = svm_model.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# prediction
y_pred = svm_model.predict(x_test)

# Evaluation
accuracy = accuracy_score(y_test, y_pred)*100
print(accuracy)
```

Classification

```python
# classification
class_report = classification_report(y_test, y_pred)
print(class_report)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn import metrics
cd=metrics.ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=[0,1,2,3])
cd.plot()
plt.show()
```
